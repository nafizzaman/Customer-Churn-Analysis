```python
print("Duplicate rows:", df.duplicated().sum())
print("\nColumns with whitespace issues:")
for col in df.columns:
    if df[col].dtype == "object":
        dirty = df[col].astype(str).str.contains(r"^\s|\s$", regex=True).sum()
        if dirty > 0:
            print(f"{col}: {dirty} entries")


df.columns[df.columns.str.contains(" ")]
```

```python
df['Total Charges'] = pd.to_numeric(df['Total Charges'].str.strip(), errors='coerce')
```

```python
df = df.dropna(subset=['Total Charges'])
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
print(df.info())
print(df.isna().sum())
```

```python
df.to_csv("../data/processed/telco_customer_churn_cleaned.csv", index=False, encoding='utf-8')
```
